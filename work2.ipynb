```python
import numpy as np
import pandas as pd
```

```python
n1,n2,list1,list2=0,0,[],[]
num1=int(input("横何マスの計算がしたいですか？"))
num2=int(input("縦何マスの計算がしたいですか？"))
array1=np.random.randint(1,10,(1,num1))
while n1<num1:
  list1.append(array1[0,n1])
  n1+=1
array2=np.random.randint(1,10,(num2,1))
while n2<num2:
  list2.append(array2[n2,0])
  n2+=1
df=pd.DataFrame(np.zeros((num2,num1)))
df.index=[list2]
df.columns=[list1]
df=df.astype("int")
display(df)
array_ans=np.zeros((num2,num1))
print("計算に入ります")
for i in range(num2):
  for j in range(num1):
    print("現在"+str(i+1)+"行"+str(j+1)+"列目")
    ans=int(input("A:"))
    array_ans[i,j]=ans
df_ans=pd.DataFrame(array_ans)
df_ans=df_ans.astype("int")
df_ans.index=[list2]
df_ans.columns=[list1]
if np.all(array_ans==array1*array2):
  print("Congratulation!!")
else:
  print("Fail...")
display(df_ans)
df_true=pd.DataFrame(array1*array2)
df_true.index=[list2]
df_true.columns=[list1]
display(df_true)
```
